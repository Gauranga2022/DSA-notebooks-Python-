Subarray of a particular length in O(n)

```python
def sub_arr(arr, s):
  l,sum,r = 0,0,0
  for i in range(len(arr)):
    while sum<s and r<len(arr):
      sum+=arr[r]
      r+=1
    if sum==s :
      return((l,r-1))
    sum-=arr[l]
    l+=1
  return(-1)
sub_arr([1 , 12 , 2 , 14 , 3 , 16 , 4],19)
```

Count the triplets

```python
def count_the_triplets(arr):
  count,tuples = 0,[]
  for i in range(len(arr)):
    arr.sort()
    third = arr[i]
    arr.remove(arr[i])
    #print(arr)
    left, right = 0, len(arr)-1
    while left<right:
      #print(arr[left],arr[right],third)
      if arr[left]+arr[right]==third:
        count+=1
        tuples.append((arr[left],arr[right],third))
        left+=1
        right-=1
      elif arr[left]+arr[right]>third:
        right-=1
      elif arr[left]+arr[right]<third:
        left+=1
    arr.append(third)
    #print(arr)
  return(count, tuples)
```

```python
count_the_triplets([5, 32, 1, 7, 10, 50, 19, 21, 2])
```

Missing number in an array in O(n)

```python
def missing(arr,N):
  dic = {}
  for i in range(N):
    dic[i]=0 #O(n)
  for j in arr:
    dic[j]+=1 #O(n)
  for k in dic.keys():
    if dic[k]==0:
      return(k) #O(n)
  return(-1)
#net: O(3n)=O(n)
```

```python
missing([1,2,0,3,5,7,6],8)
```

Merge Two **Sorted** Arrays in O((n+m)*log(n+m)) time without using extra space

```python
def merge(arr1,arr2):
  i,j,k = 0,0,len(arr1)-1
  while i<=k and j<len(arr2):
    if arr1[i]<arr2[j]:
      i+=1
    else:
      temp = arr2[j]
      arr2[j] = arr1[k]
      arr1[k] = temp
      j+=1
      k-=1
    #print(arr1,arr2)
  arr1.sort()
  arr2.sort()
  return(arr1+arr2)
```

```python
merge([1,3,5,7],[2,4,6,8,10])
```

Rearranging an array in max-min form in O(n), space complexity O(n)

```python
def sort_max_min(arr):
  left, right = 0, len(arr)-1
  result = []
  flag = True
  while left<=right:
    if flag:
      result.append(arr[right])
      right-=1
    else:
      result.append(arr[left])
      left+=1
    flag = not(flag)
  return(result)
```

```python
sort_max_min([1,2,3,4,5,6,7,8,9])
```

Count no. of inversions in O(nlogn)

```python
def inv_count(arr): #naive appproach
  count = 0
  for i in range(len(arr)):
    for j in range(i,len(arr)):
      if arr[i] > arr[j]:
        count+=1
  return(count)
```

Given an array of size N containing only 0s, 1s, and 2s; sort the array in ascending order in O(n) time and O(1) space

```python
def sort(A):
  c0,c1,c2 = 0,0,0
  for i in range(len(A)):
    if A[i]==0:
      c0+=1
    elif A[i]==1:
      c1+=1
    else:
      c2+=1
  for i in range(c0):
    A[i]=0
  for j in range(c0,c0+c1):
    A[j]=1
  for k in range(c0+c1,c0+c1+c2):
    A[k]=2
  return(A)
```

```python
sort([0,1,1,2,0,1,2,1,0])
```

The task is to complete the function equilibriumPoint() which takes the array and n as input parameters and returns the point of equilibrium.

Expected Time Complexity: O(n)
Expected Auxiliary Space: O(1)

```python
def equilibrium(arr):
  leftsum, totalsum = 0,0
  for i in range(len(arr)):
    totalsum+=arr[i]
  for i in range(len(arr)):
    rightsum = totalsum-leftsum-arr[i]
    if leftsum==rightsum:
      return(i)
    leftsum+=arr[i]
  return(-1)
```

```python
equilibrium([7,2,3,4,5])
```

Given an array A of positive integers. Your task is to find the leaders in the array. An element of array is leader if it is greater than or equal to all the elements to its right side. The rightmost element is always a leader.

Expected Time Complexity: O(n)
Expected Auxiliary Space: O(n)

```python
def leader(arr):
  lead,n,max_so_far = [arr[-1]],len(arr),arr[-1]
  for i in range(n-1,0,-1):
    if arr[i]>max_so_far:
      max_so_far=arr[i]
      lead.append(arr[i])
  return(lead)
```

Given arrival and departure times of all trains that reach a railway station. Find the minimum number of platforms required for the railway station so that no train is kept waiting.
Consider that all the trains arrive on the same day and leave on the same day. Arrival and departure time can never be the same for a train but we can have arrival time of one train equal to departure time of the other. At any given instance of time, same platform can not be used for both departure of a train and arrival of another train. In such cases, we need different platforms.
Expected Time Complexity: O(nLogn)
Expected Auxiliary Space: O(n)



```python
def min_platform(arr,dep):
  n = len(arr)
  arr.sort()
  dep.sort()
  i,j,platform,result=1,0,1,1
  while i<n and j<n:
    if arr[i]<dep[j]:
      i+=1
      platform+=1
    elif arr[i]>dep[j]:
      platform-=1
      j+=1
    if platform>result:
      result=platform
  return(result)
```

Given an array arr[] of positive integers of size N. Reverse every sub-array group of size K.

Note: If at any instance, there are no more subarrays of size greater than or equal to K, then reverse the last subarray (irrespective of its size). You shouldn't return any array, modify the given array in-place.

Expected Time Complexity: O(N)
Expected Auxiliary Space: O(1)

```python
def reverse_in_group(arr,k):
  i,j=0,k-1
  #print(len(arr)//k)
  for n in range((len(arr)//k)):
    while i<j:
      arr[i],arr[j]=arr[j],arr[i]
      i+=1
      j-=1
    #print(arr)
    i,j=0,k-1
    i+=(n+1)*k
    j+=(n+1)*k
  j=len(arr)-1
  #print(i,j)
  while i<j:
    arr[i],arr[j]=arr[j],arr[i]
    i+=1
    j-=1
  return(arr)
```

Given an array arr[] of N non-negative integers representing the height of blocks. If width of each block is 1, compute how much water can be trapped between the blocks during the rainy season.
Expected Time Complexity: O(N)
Expected Auxiliary Space: O(N)

```python
def trap(arr):
  max_last = arr[0]
  sum = 0
  for i in range(len(arr)):
    if arr[i] <= max_last:
      sum += (max_last-arr[i])
    else:
      max_last = arr[i]
  return(sum)
```

```python
trap([7,4,0,9])
```

Given an array arr of N integers, write a function that returns true if there is a triplet (a, b, c) that satisfies a2 + b2 = c2, otherwise false. Expected Time Complexity: O(max(Arr[i])^2)
Expected Auxiliary Space: O(max(Arr[i]))

```python
def checkTriplet(arr):
  sq = []
  for i in range(len(arr)):
    sq.append(arr[i]**2)
  sq.sort()
  flag,k = False,len(arr)-1
  while k>1:
    i,j,pointer=0,k-1,k
    while (i<j):
      if sq[i]+sq[j]<sq[pointer]:
        i+=1
      elif sq[i]+sq[j]>sq[pointer]:
        j-=1
      elif sq[i]+sq[j]==sq[pointer]:
        flag = True
        break
    k-=1
  return flag
```

Given an array A[ ] of positive integers of size N, where each value represents the number of chocolates in a packet. Each packet can have a variable number of chocolates. There are M students, the task is to distribute chocolate packets among M students such that :
1. Each student gets exactly one packet.
2. The difference between maximum number of chocolates given to a student and minimum number of chocolates given to a student is minimum.

Expected Time Complexity: O(N*Log(N))
Expected Auxiliary Space: O(1)

```python
def birthdaychoco(arr,m):
  arr.sort()
  n = len(arr)
  min_diff = arr[m-1]-arr[0]
  for i in range(n-m+1):
    diff = arr[i+m-1]-arr[i]
    min_diff = min(diff, min_diff)
  return(min_diff)
```

```python
birthdaychoco([3, 4, 1, 9, 56, 7, 9, 12],5)
```
