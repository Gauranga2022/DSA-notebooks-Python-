String Matching Brute Force

```python
def stringmatchrl(t,p): #left to right
  poslist = []
  for i in range(len(t)-len(p)+1):
    matched = True
    j = 0
    while j < len(p) and matched == True:
      if t[i+j]!=p[j]:
        matched = False
      j+=1
    if matched:
      poslist.append(i)
  return(poslist)
```

```python
stringmatchrl('banana','ba')
```

```python
def stringmatchlr(t,p): #right to left
  poslist = []
  for i in range(len(t)-len(p)+1):
    matched = True
    j = len(p)-1
    while j >=0 and matched == True:
      if t[i+j]!=p[j]:
        matched = False
      j-=1
    if matched:
      poslist.append(i)
  return(poslist)
```

```python
stringmatchlr('banana','ba')
```

Boyer Moore Algorithm

```python
def Boyermoore(t,p):
  poslist = []
  last = {}
  for i in range(len(p)):
    last[p[i]]=i
  #print(last)
  i=0
  while i < len(t)-len(p)+1:
    j = len(p)-1
    matched = True
    while j >= 0 and matched:
      if t[i+j]!= p[j]:
        matched = False
      j-=1
      #print(matched)
    if matched:
      i+=1
      poslist.append(i)
    else:
      j+=1
      if t[i+j] in last.keys():
        i=i+max(1,j-last[t[i+j]])
      else:
        i+=i+j+1
  return(poslist)
```
