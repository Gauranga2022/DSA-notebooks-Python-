Longest Common Subsequnce

```python
def lcs(u,v):
  import numpy as np
  (m,n) = (len(u),len(v))
  lcs = np.zeros((m+1,n+1)) #(m+1,n+1) for base case
  for i in range(m-1,-1,-1):
    for j in range(n-1,-1,-1):
      if u[i]==v[j]:
        lcs[i,j] = 1+lcs[i+1,j+1]
      else:
        lcs[i,j] = max(lcs[i,j+1],lcs[i+1,j])
  return(lcs[0,0],lcs)
```

```python
lcs('secrety','secretary')
```

Longest Common Word

```python
def lcw(u,v):
  import numpy as np
  (m,n)=(len(u),len(v))
  lcw = np.zeros((m+1,n+1))
  maxlcw = 0
  for i in range(m-1,-1,-1):
    for j in range(n-1,-1,-1):
      if u[i]==v[j]:
        lcw[i,j]=1+lcw[i+1,j+1]
      else:
        lcw[i,j]=0
      if lcw[i,j]>maxlcw:
        maxlcw = lcw[i,j]
  return(lcw,maxlcw)
```

```python
lcw('secret','secretary')
```

```python
lcw('happy','apple')
```

Longest Decreasing Sequence

```python
user_input = input()
numbers = user_input.split()
numbers = [int(num) for num in numbers]
def lds(seq):
  L = [0]*len(seq)
  L[0]=1
  path_dic = {}
  for i in range(len(seq)):
    path_dic[seq[i]]=[]
  for i in range(len(seq)):
    for j in range(i):
      if L[j]>L[i] and seq[j]>seq[i]:
        L[i]=L[j]
        path_dic[seq[i]].append(seq[j])
    L[i]+=1
    path_dic[seq[i]].append(seq[i])
    #print('L=',L)
  return(path_dic[seq[L.index(max(L))]])
print(lds(numbers))
```

```python
M = eval(input())
(x1, y1) = eval(input())
(x2, y2) = eval(input())

def MaxCoinPath(M, x1, y1, x2, y2):
    if x1 >= len(M) or y1 >= len(M[0]) or x2 >= len(M) or y2 >= len(M[0]):
        return 0
    if x2 == x1 and y2 == y1:
        return M[x1][y1]
    else:
        return M[x1][y1] + max(
            MaxCoinPath(M, x1 + 1, y1, x2, y2),
            MaxCoinPath(M, x1, y1 + 1, x2, y2)
        )

print(MaxCoinPath(M, x1, y1, x2, y2))
```
