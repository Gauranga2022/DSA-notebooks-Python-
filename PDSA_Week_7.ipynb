# Max Overlap GRPA 2

```python
def overlap(x,y):
  return((x[1]<=y[1]<=x[2]<=y[2]) or (y[1]<=x[1]<=y[2]<=x[2]) or (y[1]<=x[1]<=x[2]<=y[2]))
def overlap_no(l,x):
  count = 0
  for i in range(len(l)):
    if overlap(x,l[i]):
      count+=1
  return(count)
def minimum_platform(l):
  arr = sorted(l, key=lambda x: x[1])
  res = 0
  for i in range(len(l)):
    res = max(res, overlap_no(l,l[i]))
  return(res)
```
